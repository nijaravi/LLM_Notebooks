Writing To CSV

```python
import tabula
import pandas as pd

# Define the PDF file path
pdf_path = "/content/2_table.pdf"

# Read the PDF and extract tables
dfs = tabula.read_pdf(pdf_path, pages='1', multiple_tables=True)

# Check if any tables were extracted
if dfs:
    for i, df in enumerate(dfs):
        # Save each table to a separate CSV file
        df.to_csv(f"table{i+1}.csv", index=False)

    print(f"{len(dfs)} CSV files saved successfully.")
else:
    print("No tables found in the PDF.")
```

Writing To Excel

```python
import tabula
import pandas as pd

# Define the PDF file path
pdf_path = "/content/2_table.pdf"

# Read the PDF and extract tables
dfs = tabula.read_pdf(pdf_path, pages='1')

if len(dfs) > 0:
    # Create a Pandas Excel writer
    excel_writer = pd.ExcelWriter("tables.xlsx")

    for i, df in enumerate(dfs):
        sheet_name = f'Table{i + 1}'  # Generate a unique sheet name for each table
        df.to_excel(excel_writer, sheet_name=sheet_name, index=False)

    # Save and close the Excel writer
    excel_writer.save()

    print(f"Excel file with {len(dfs)} sheets saved successfully.")
else:
    print("No tables found in the PDF.")
```
